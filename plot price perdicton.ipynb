```python
import numpy as np
from sklearn.linear_model import LinearRegression

# # Create the data
# square_footage = np.array([1000, 1500, 2000, 2500, 3000])
# price = np.array([200000, 300000, 400000, 500000, 600000]).reshape(-1, 1)
# Create the data
square_footage = np.array([1000, 1500, 2000, 2500, 3000])
price = np.array([200000, 300000, 350000, 380000, 500000]).reshape(-1, 1)


square_footage = square_footage.reshape(-1, 1)



# Create the linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(square_footage, price)

# Print the coefficients
print(model.coef_)
print(model.intercept_)
# Predict the price of a house with 2500 square feet
prediction = model.predict(np.array([2800]).reshape(-1, 1))
print("prediction: ", prediction)
```

```python
#  for one featuer
import matplotlib.pyplot as plt
# Plot the data
plt.scatter(square_footage, price)
plt.plot(square_footage, model.predict(square_footage), color='red')
plt.xlabel('Square footage')
plt.ylabel('Price')
plt.show()
```
